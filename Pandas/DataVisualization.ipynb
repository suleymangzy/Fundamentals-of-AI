# Veri Görselleştirme

Pandas, verilerimiz görselleştirmemiz için çeşitli araçlar ve fonksiyonlar sunar. Pandas' ın bazı görselleştirme yetenekleri aşağıdaki gibidir.

## Çizgi Grafiği (Line Plot)

Çizgi grafiği, verileri çizgi üzerinde noktalar ile temsil eden bir grafik türüdür. Bu grafik türü genellikle zamana bağlı değişimleri veya sürekli bir değişkenin eğilimini göstermek için kullanılır.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

df = pd.DataFrame({'x': x, 'y': y})

df.plot(x='x', y='y', kind='line', color='orange')

plt.title('Line Plot')
plt.xlabel('x values')
plt.ylabel('y values')

plt.show()
```

Yukarıdaki örnekte, x ve y listelerinden bir veri çerçevesi oluştururuz.  Ardından, `plot()` fonksiyonunu kullanarak çizgi grafiğini oluşturduk. `x` ve `y` parametreleri, hangi sütunların `x` ve `y` eksenlerinde kullanacağımızı belirtir. `kind=line` parametresi, çizgi grafiğini oluşturmak için kullanılır.

Grafik özelliklerini belirtmek için `plt.title()`, `plt.xlabel()` ve `plt.ylabel()` fonksiyonlarını kullanırız. Son olarak, `plt.show()` fonksiyonu ile grafiği görüntüleriz.

## Sütun Grafiği (Bar Plot)

Sütun grafiği, kategorik verilerin farklı gruplar veya kategoriler arasındaki karşılaştırmalrını göstermek için kullanılan bir grafik türüdür. Her kategori sütunlar kullanılarak verilerin görsel temsili sağlanır.

```python
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [20, 10, 15, 25]

df = pd.DataFrame({'Category': categories, 'Value': values})

df.plot(x='Category', y='Value', kind='bar', color='red')

plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')

plt.show()
```

## Histogram Grafiği

Histogram, sürekli bir değişkenin dağılımını göstermek için kullanılan bir grafik türüdür. Değişkenin değer aralıkları (bins) belirlenir ve her bir arlıkta kaç örnek olduğu görsel olarak temsil edilir. Histogram, veri dağılımının genel şeklini, merkeszi eğilimini ve yayılımını görmek için kullanılır.

```python
data = [1, 1, 1, 2, 2, 3, 3, 3, 3, 4, 4, 5, 6, 6, 7, 7, 7, 8, 8, 8]

df = pd.DataFrame({'Data': data})

counts, bins, _ = plt.hist(df['Data'], bins=8, color='orange', alpha=0.3)

plt.plot(bins[:-1], counts, color='red', marker='o')

plt.title('Histogram Plot')
plt.xlabel('Data')
plt.ylabel('Frequency')

plt.show()
```

## Nokta Grafiği (Scatter Plot)

Nokta grafiği, iki sürekli değişken arasındaki ilişkiyi veya dağılımı görselleştirmek için kullanılan bir grafik türüdür. Her bir veri noktası, bir eksen üzerinde ise diğer değişkenin değerini temsil eder. Nokta grafiği, değişkenler arasındaki ilişki analiz etmek, korelasyonu görmek veya aykırı değerleri tespit etmek için kullanılır.

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 9]

df = pd.DataFrame({'x': x, 'y': y})

df.plot(x='x', y='y', kind='scatter', color='orange')
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
```

## Kutu Grafiği (Box Plot)

Kutu grafiği, verilerin merkezi eğilimini, yayılımını ve aykırı değerleri görsdelleştirmek için kullanılan bir grafik türüdür. Kutu grafiği, verilerin minimum, birinci çeyrek (Q1), medyan (Q2), üçüncü çeyrek (Q3) ve maksimum değerlerini temsil eden bir kutu ve bu kutunun üzerinde ve altındda yer alan `bıyıklar (whisker)` ile gösterilir.

```python
data = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55]

df = pd.DataFrame({'Data': data})

df.plot.box()

plt.title('Box Plot')
plt.ylabel('Data')

plt.show()
```

## Pasta Grafiği (Pie Plot)

Pasta grafiği, bir bütünün farklı parçalarının oranlarını görselleştirmek için kullanılan bir grafik türüdür. Pasta grafiği, verilerin yüzdesel dağılımını göstermek için kullanılır. Her bir veri noktası, bir dilim olarak temsil edilir ve tüm dilimlerin toplamı %100' e eşittir.

```python
sizes = [30, 25, 20, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']

colors = plt.cm.tab10.colors

plt.pie(
    sizes,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',  # yüzde değerleri yaz
    startangle=90,  # grafiği 90° den başlat
    counterclock=False  # saat yönünde sıralasın
)

plt.title('Pie Plot')

plt.show()
```

## Alan Grafiği (Area Plot)

Alan grafiği, bir değişkenin zamana veya başka bir bağımsız değişkene bağlı olarak nasıl değiştiğini görselleştirmek için kullanılan bir grafik türüdür. Alan grafiği, değişkenin değelerinin zaman veya başak bir eksen boyunca nasıl yayıldığını ve birbirini örttüğünü gösterir. Bu grafik türü, değişkenin toplamı veya yüzdesel katkıları gibi bir bütün içindeki bileşenlerin gösterimi için de kullanılabilir.

```python
years = [2010, 2011, 2012, 2013, 2014]
data1 = [10, 15, 12, 8, 20]
data2 = [5, 8, 6, 10, 12]

df = pd.DataFrame({'Year': years, 'Data1': data1, 'Data2': data2})

df.plot.area(x='Year')

plt.title('Area Plot')
plt.xlabel('Year')
plt.ylabel('Data')

plt.show()
```

## Yoğunluk Grafiği (Density Plot)

Yoğunluk grafiği, veri dağılımının genel şeklini ve yoğunluk tahminini görselleştirmek için kullanılan bir grafik türüdür. Yoğunluk grafiği, veri setinin olasılık yoğunluğunu yansıtan bir eğri şeklinde çizilir.

Yoğunluk grafiğini oluşturmak için `Saborn` kütüphanesini kullanırız.

```python
import seaborn as sns
```

```python
data = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55]

df = pd.DataFrame({'Data': data})

sns.kdeplot(data=df['Data'].to_numpy())

plt.title('Density Plot')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

## İleri Düzey Pandas

### Zaman Serileri Analizi

Zaman serileri analizi, zamanla değişen verilerin analiz ve tahminini içeren istatistiksel bir yöntemdir. Zaman serileri genellikle düzenli zaman aralıklarında ölçülen verilerdir ve önceki değerlerin gelecekteki değerleri tahmin etmek için kullanılabileceği bir yapıya sahiptir. Pandas kütüphanesi, zaman serileri analizi için bir dizi özel işlev ve veri yapısı sağlar.

```python
data = {'Tarih': ['2021-01-01', '2021-01-02', '2021-01-03', '2021-01-04',
                  '2021-01-05'],
        'Değer': [10, 15, 12, 18, 20]}

df = pd.DataFrame(data)

df['Tarih'] = pd.to_datetime(df['Tarih'], format='%Y-%m-%d')
df.set_index('Tarih', inplace=True)

df.plot()

plt.title('Time Series')
plt.xlabel('History')
plt.ylabel('Value')

plt.show()
```

Yukarıdaki örnek kodda, bir zaman serisi veri seti oluşturuyoruz. `Tarih` sütununda `pd.to_datetime()` fonksiyonuyla zaman formatına dönüştürüyoruz. Ardından, `set_index()` fonksiyonunu kullanarak `Tarih` sütununu indeks olarak ayarlıyoruz. Daha sonra, `df.plot()` ile zaman serisi veriyi çiziyoruz ve `plt` kullanarak etiketleri ve başlığı ekliyoruz.

### Çoklu İndeksleme ve Çok Seviyeli Veri Çerçeveleri

Çoklu indeksleme veya çok seviyeli indeksleme, veri çerçevelerinde birden fazla sütunu indeks olarak kullanma yöntemidir. Bu şekilde, verileri daha karmaşık bir yapıda organize etmek ve daha etkili bir şekilde işlemek mümkündür. Çok sseviyeli veri çerçeveleri, genellikle hiyerarşik veya ilişklili veri yapılarını temsil etmek için kullanılır.

```python
data = {'Şehir': ['İstanbul', 'İstanbul', 'İstanbul', 'Ankara', 'Ankara',
                  'Ankara'],
        'Bölge': ['Avrupa', 'Avrupa', 'Anadolu', 'Anadolu', 'Anadolu',
                  'Anadolu'],
        'Mevsim': ['Yaz', 'Kış', 'İlkbahar', 'Kış', 'Yaz', 'İlkbahar'],
        'Sıcaklık': [30, 10, 20, 5, 25, 15]}
df = pd.DataFrame(data)
df.set_index(['Şehir', 'Bölge', 'Mevsim'], inplace=True)

print('DataFrame: \n', df)
```

### Büyük Veri Kümeleriyle Çalışmak için Performans İyileştirmeleri

#### Veri Okuma ve Yazma Performansını İyileştirme

* Daha hızlı veri formatları kullanabiliriz. Öörneğin, Parqute veya Feather gibi optimize edilmiş dosya formatlarını tercih edebiliriz.
* Veri kümesini paralel olarak okuyabiliriz. Örneğin, Dask veya Modin gibi araçları kullanarak verileri paralel olarak işleyebiliriz.
* Veri kümesini parçalara bölebilir ve her bir parçayı ayrı ayrı işleyebilirz. Bu, işlem süresini azaltabilir.

#### Bellek Kullanımı Optimizasyonu

* Gereksiz sütunları veya satırları atabiliriz. Bu şekilde bellek kullanımını azaltabiliriz.
* Veri tiplerini optimize edebiliriz. Örneğin, daha küçük veri tipleri kullanarak bellek alanında tasarruf  sağlayabiliriz.

#### Paralel İşleme ve Dağıtık Hesaplama

* Büyük veri kümelerini paralel olarak işleyebiliriz. Örneğin, Spark veya Hadoop gibi araçları kullanarak paralel hesaplama yapabiliriz.
* Verileri birden çok makine veya küme üzerinde dağıtabiliriz. Bu şekilde işlem gücünü ölçeklendirebiliriz.
